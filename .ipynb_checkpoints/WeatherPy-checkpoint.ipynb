```python
# Dependencies and Setup
import json
import seaborn as sns
import requests as req
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random
from citipy import citipy
import time
import pprint

api_key = "87ab869df515895db58f8a9d1139ae61"
url = "http://api.openweathermap.org/data/2.5/weather?"
# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
columns = ["City"]
columnname = ["City", "Country","Temperature (F)",
                "Latitude","Longitude","Humidity (%)",
                "Cloudiness (%)", "Wind Speed (mph)"]
weatherdf = pd.DataFrame(columns=columnname)
count = 1

for city in cities:
    try:
            # Build query URL
            query_url = url + "appid=" + api_key + "&q=" + city + "&units=imperial"

            # Get weather data
            weather_response = req.get(query_url).json()
            cityName = weather_response["name"]
            weatherdf.set_value(count, "City", weather_response["name"])
            weatherdf.set_value(count, "Country", weather_response["sys"]["country"])
            weatherdf.set_value(count, "Temperature (F)", weather_response["main"]["temp"])
            weatherdf.set_value(count, "Latitude", weather_response["coord"]["lat"])
            weatherdf.set_value(count, "Longitude", weather_response["coord"]["lon"])
            weatherdf.set_value(count, "Humidity (%)", weather_response["main"]["humidity"])
            weatherdf.set_value(count, "Cloudiness (%)", weather_response["clouds"]["all"])
            weatherdf.set_value(count, "Wind Speed (mph)", weather_response["wind"]["speed"])
            print("City Number: " + str(count) + ", City Name: " + cityName)

            count += 1
            sleep = 1
    except KeyError:
        print("Missing result, skipping")

    

```
